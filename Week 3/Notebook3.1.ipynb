## Random Walk

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
import time

# Seed for reproducibility
random.seed(datetime.datetime.now().strftime('%s'))

def getLinks(articleUrl):
    try:
        html = urlopen(f'https://en.wikipedia.org{articleUrl}')
        bs = BeautifulSoup(html, 'html.parser')
        time.sleep(1)  # Be polite to Wikipedia
        return bs.find('div', {'id': 'bodyContent'}).find_all(
            'a', href=re.compile('^(/wiki/)((?!:).)*$'))
    except Exception as e:
        print(f"Error accessing {articleUrl}: {e}")
        return []

def randomWalk(articleUrl, current_depth, max_depth):
    if current_depth > max_depth:
        print("Reached max depth. Stopping.")
        return

    links = getLinks(articleUrl)
    if not links:
        print("No more links found. Stopping.")
        return

    newArticle = random.choice(links).attrs['href']
    print(f"[Depth {current_depth}] -> {newArticle}")
    randomWalk(newArticle, current_depth + 1, max_depth)

# Start the random walk from a known article
randomWalk('/wiki/Kevin_Bacon', current_depth=0, max_depth=5)
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time

visited_pages = set()

def get_links(page_url, depth, max_depth=2):
    if depth > max_depth:
        return

    full_url = f'https://en.wikipedia.org{page_url}'
    print(f"\n[Depth {depth}] Visiting: {full_url}")

    try:
        html = urlopen(full_url)
        bs = BeautifulSoup(html, 'html.parser')
        time.sleep(1)  # Throttle requests to avoid overloading the server

        # Print page title
        title = bs.h1.get_text() if bs.h1 else "No Title"
        print(f"Page Title: {title}")

        # Print first paragraph if available
        body_paragraphs = bs.find('div', {'id': 'bodyContent'}).find_all('p')
        if body_paragraphs:
            print("Intro:", body_paragraphs[0].get_text(strip=True))

        # Optional: print edit link if available
        edit_link = bs.find(id='ca-edit')
        if edit_link:
            print("Edit Link:", edit_link.find('a').attrs['href'])

    except Exception as e:
        print(f"Error accessing {full_url}: {e}")
        return

    # Now crawl linked pages
    links = bs.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))
    for link in links:
        href = link.attrs['href']
        if href not in visited_pages:
            visited_pages.add(href)
            get_links(href, depth + 1, max_depth)

# Start crawling from a specific Wikipedia page (e.g., Python programming language)
get_links('/wiki/Python_(programming_language)', depth=0, max_depth=2)
```

## Collecting Data Across an Entire Site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time

# Track visited pages
visited_pages = set()

# Crawl limit
def getLinks(pageUrl, depth, max_depth=2):
    if depth > max_depth:
        return

    full_url = f'https://en.wikipedia.org{pageUrl}'
    print(f"\n[Depth {depth}] Visiting: {full_url}")

    try:
        html = urlopen(full_url)
        bs = BeautifulSoup(html, 'html.parser')
        time.sleep(1)  # Be nice to Wikipedia

        # Page Title
        title = bs.h1.get_text() if bs.h1 else "No Title"
        print(f"Title: {title}")

        # First paragraph (if available)
        body_paragraphs = bs.find('div', {'id': 'bodyContent'}).find_all('p')
        if body_paragraphs:
            print("Intro:", body_paragraphs[0].get_text(strip=True))

        # Edit link (if exists)
        edit_link = bs.find(id='ca-edit')
        if edit_link:
            print("Edit Link:", edit_link.find('a').attrs['href'])

    except Exception as e:
        print(f"Error on page {pageUrl}: {e}")
        return

    # Look for more internal links
    links = bs.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))
    for link in links:
        href = link.attrs['href']
        if href not in visited_pages:
            visited_pages.add(href)
            getLinks(href, depth + 1, max_depth)

# Start crawling from this page
getLinks('/wiki/General-purpose_programming_language', depth=0, max_depth=2)
```

## Crawling across the Internet

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import random
import time

# Get all internal links on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        href = link.attrs.get('href')
        if not href:
            continue
        parsed = urlparse(href)
        if parsed.netloc == '':
            clean_url = f'{scheme}://{netloc}/{href.strip("/")}'
            internalLinks.add(clean_url)
        elif parsed.netloc == netloc:
            internalLinks.add(href)
    return list(internalLinks)

# Get all external links on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        href = link.attrs.get('href')
        if not href:
            continue
        parsed = urlparse(href)
        if parsed.netloc and parsed.netloc != netloc:
            externalLinks.add(href)
    return list(externalLinks)

# Get a random external link or fall back to internal ones
def getRandomExternalLink(startingPage):
    try:
        print(f"\nVisiting: {startingPage}")
        html = urlopen(startingPage, timeout=10)
        bs = BeautifulSoup(html, 'html.parser')
        time.sleep(1)

        externalLinks = getExternalLinks(bs, startingPage)
        if externalLinks:
            return random.choice(externalLinks)

        print("No external links found. Trying internal links...")
        internalLinks = getInternalLinks(bs, startingPage)
        if internalLinks:
            return getRandomExternalLink(random.choice(internalLinks))
        else:
            print("No internal links found either.")
            return None

    except Exception as e:
        print(f"Failed to access {startingPage}: {e}")
        return None

# Recursive external-only crawler with depth control
def followExternalOnly(startingSite, max_depth=5, current_depth=0):
    if current_depth >= max_depth:
        print("Reached max depth. Stopping crawl.")
        return

    externalLink = getRandomExternalLink(startingSite)
    if externalLink:
        print(f"[Step {current_depth}] -> Following: {externalLink}")
        followExternalOnly(externalLink, max_depth, current_depth + 1)
    else:
        print("No further external links to follow.")

# Start the crawl
followExternalOnly('https://www.oreilly.com/', max_depth=5)
```

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random

#Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

#Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    bs = BeautifulSoup(urlopen(startingPage), 'html.parser')
    externalLinks = getExternalLinks(bs, startingPage)
    if not len(externalLinks):
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        return getRandomExternalLink(random.choice(internalLinks))
    else:
        return random.choice(externalLinks)

def followExternalOnly(startingSite):
    externalLink = getRandomExternalLink(startingSite)
    print(f'Random external link is: {externalLink}')
    followExternalOnly(externalLink)


followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
# Collects a list of all external URLs found on the site
allExtLinks = []
allIntLinks = []


def getAllExternalLinks(url):
    bs = BeautifulSoup(urlopen(url), 'html.parser')
    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)
    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.append(link)
            print(link)

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.append(link)
            getAllExternalLinks(link)


allIntLinks.append('https://oreilly.com')
getAllExternalLinks('https://www.oreilly.com/')
```
